```python
# Function to perform text cleaning without removing stop words
def clean_text(text):
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)

    # Remove special characters and numbers
    text = re.sub('[^a-zA-Z]', ' ', text)

    # Remove specific words
    text = text.replace('div', '').replace('li', '').replace('ul', '')

    # Remove extra white spaces
    text = re.sub(r'\s+', ' ', text).strip()

    # Convert text to lowercase
    text = text.lower()

    return text
```

```python
# Clean columns
merged_df['title'] = merged_df['title'].apply(clean_text)
```
